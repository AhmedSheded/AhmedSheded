```python
import tensorflow_hub as hub
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
model = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

```python
def load_image(img_path):
  img = tf.io.read_file(img_path)
  img = tf.image.decode_image(img, channels=3)
  img = tf.image.convert_image_dtype(img, tf.float32)
  img = img[tf.newaxis, :]
  return img
```

```python
img = load_image('photo.jpg')
style_img = load_image('p.jpg')
arya = load_image('arya.jpg')
```

```python
plt.imshow(np.squeeze(img))
plt.show()
```

```python
plt.imshow(np.squeeze(style_img))
plt.show()
```

```python
stylized_image = model(tf.constant(img), tf.constant(style_img))[0]
```

```python
plt.imshow(np.squeeze(stylized_image))
plt.show()
```

```python
cv2.imwrite('genrated.jpg', cv2.cvtColor(np.squeeze(stylized_image)*255, cv2.COLOR_BGR2RGB))
```
